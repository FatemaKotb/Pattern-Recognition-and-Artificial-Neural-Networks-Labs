### 1. Line Plots

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting
plt.style.use('dark_background')            # 'default' is the default
plt.figure(figsize=(5, 3), dpi=150)
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

More general example

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting both sin and cos functions on the same axes
plt.figure(figsize=(8, 3), dpi=150)
plt.plot(x, y1, 'r-', label='sin(x)')
plt.plot(x, y2, 'b-', label='cos(x)')
plt.title('Sin and Cos Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()  # Show legend
plt.show()
```

#### 3D Version

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z1 = np.sin(np.sqrt(x**2 + y**2))
z2 = np.cos(np.sqrt(x**2 + y**2))

# Creating a 3D plot
fig = plt.figure(figsize=(10, 6), dpi=140)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z1, cmap='viridis', alpha=0.5)  # Plot the first surface
ax.plot_surface(x, y, z2, cmap='plasma', alpha=0.5)   # Plot the second surface

ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis', labelpad=1)

# Remove grid
ax.grid(False)
plt.show()
```

### 2. Scatter Plots

#### Scatter plots in 2D

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data
num_points = 1000
x = np.random.rand(num_points, 2)        # Generating 2D vectors for x
y = np.random.randint(0, 3, num_points)  # Generating random integers for y in range [0, 4)

# Mapping colors to each class
colors = ['y', 'r', 'b']
class_labels = [f'Class {i}' for i in range(4)]

# Creating scatter plot (can use a for loop of course)
plt.figure(figsize=(12, 5), dpi=100)
plt.scatter(x[y == 0, 0], x[y == 0, 1], c=colors[0], label=class_labels[0])
plt.scatter(x[y == 1, 0], x[y == 1, 1], c=colors[1], label=class_labels[1])
plt.scatter(x[y == 2, 0], x[y == 2, 1], c=colors[2], label=class_labels[2])

plt.title('Scatter Plot with 2D Vectors')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

#### Scatter plots in 3D

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data
num_points = 100
x = np.random.rand(num_points, 3)  # Generating 3D vectors for x
y = np.random.randint(0, 4, num_points)  # Generating random integers for y in range [0, 4)

# Mapping colors to each class
colors = ['r', 'g', 'b', 'y']
class_labels = [f'Class {i}' for i in range(4)]

# Creating scatter plot
fig = plt.figure(figsize=(8, 6), dpi=140)
ax = fig.add_subplot(111, projection='3d')

for i in range(4):
    ax.scatter(x[y == i, 0], x[y == i, 1], x[y == i, 2], c=colors[i], label=class_labels[i])

ax.set_title('3D Scatter Plot with 3D Vectors')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
ax.legend()

# Remove grid
ax.grid(False)
# ax.xaxis.pane.set_facecolor('black') to change plane color
plt.show()
```

### 3. Other Plots

#### Bar Plot

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

# Plotting
fig = plt.figure(figsize=(4, 3), dpi=140)
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

#### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(1000)

# Plotting
fig = plt.figure(figsize=(4, 3), dpi=140)
plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(100)

# Plotting
fig = plt.figure(figsize=(4, 3), dpi=140)
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

#### Pie Chart

```python
import matplotlib.pyplot as plt

# Sample data
sizes = [20, 30, 25, 25]
labels = ['A', 'B', 'C', 'D']

# Plotting
fig = plt.figure(figsize=(4, 3), dpi=140)
plt.pie(sizes, labels=labels)
plt.title('Pie Chart')
plt.show()
```

### 4. Grid of Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8), dpi=150)
# treat each axs just like plt

# Plot a bar chart
axs[0, 0].bar(['A', 'B', 'C', 'D'], [3, 7, 2, 5])
axs[0, 0].set_title('Bar Chart')

# Plot a histogram
axs[0, 1].hist(x, bins=30, color='skyblue', edgecolor='black')
axs[0, 1].set_title('Histogram')

# Plot a box plot
axs[1, 0].boxplot(x)
axs[1, 0].set_title('Box Plot')

# Plot a scatter plot
axs[1, 1].scatter(x, y)
axs[1, 1].set_title('Scatter Plot')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
